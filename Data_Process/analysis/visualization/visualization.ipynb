```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("../../preprocess/data.csv")
data
```

```python
def process_big_category(d:str):
    if d == "后端开发":
        return 0
    elif d == "移动开发":
        return 1
    elif d == "前端开发":
        return 2
    elif d == "人工智能":
        return 3
    elif d == '游戏':
        return 4
    elif d == "测试":
        return 5
    elif d == "运维/技术支持":
        return 6
    elif d == '数据':
        return 7
    
data['big_category']=data.apply(lambda x:process_big_category(x['big_category']), axis=1)
data
```

```python
def process_city(d:str):
    if d == "上海":
        return 0
    elif d == "深圳":
        return 1
    elif d == "广州":
        return 2
    elif d == "北京":
        return 3
    elif d == '重庆':
        return 4
    elif d == "成都":
        return 5
    elif d == "杭州":
        return 6
    
data['city']=data.apply(lambda x:process_city(x['city']), axis=1)
data
```

```python
data["company_type"].value_counts()
```

```python
def process_company_type(d:str):
    if d == "民营公司":
        return 0
    elif d == "上市公司":
        return 1
    elif d == "合资":
        return 2
    elif d == "国企":
        return 3
    elif d == '外资（非欧美）':
        return 4
    elif d == "外资（欧美）":
        return 5
    elif d == "创业公司":
        return 6
    elif d == "事业单位":
        return 7
    elif d == "非营利组织":
        return 8
    elif d == "外企代表处":
        return 9
    elif d == "政府机关":
        return 10
    
data['company_type']=data.apply(lambda x:process_company_type(x['company_type']), axis=1)
data
```

```python
data["company_size"].value_counts()
```

```python
def process_company_size(d:str):
    if d == "50-150人":
        return 0
    elif d == "150-500人":
        return 1
    elif d == "1000-5000人":
        return 2
    elif d == "少于50人":
        return 3
    elif d == '500-1000人':
        return 4
    elif d == "10000人以上":
        return 5
    elif d == "5000-10000人":
        return 6
    
data['company_size']=data.apply(lambda x:process_company_size(x['company_size']), axis=1)
data
```

```python
data=data.corr()
```

```python
plt.subplots(figsize=(9,9),dpi=1080,facecolor='w')# 设置画布大小，分辨率，和底色

sns.heatmap(data,annot=True, vmax=1, square=True, cmap="Blues", fmt='.2g')#annot为热力图上显示数据；fmt='.2g'为数据保留两位有效数字,square呈现正方形，vmax最大值为1
plt.show()
```
