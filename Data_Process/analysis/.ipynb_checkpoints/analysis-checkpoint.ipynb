```python
temp = data["salary"].value_counts().to_frame()
temp
```

```python
# for index, row in temp.itertuples():
#     print("{value: " + str(row) + ", name: " + "'" + index + "'" + "}," )
```

```python
a=0
b=0
c=0
d=0
e=0
f=0
for index, row in temp.itertuples():
    if float(index) < 1:
        a = a + row
    elif float(index) < 2:
        b = b + row
    elif float(index) < 3:
        c = c + row
    elif float(index) < 4:
        d = d + row
    elif float(index) < 5:
        e = e + row
    else:
        f = f + row
```

```python
print("{value: " + str(a) + ", name: " + "'" + "<1万/月" + "'" + "}," )
print("{value: " + str(b) + ", name: " + "'" + "1-2万/月" + "'" + "}," )
print("{value: " + str(c) + ", name: " + "'" + "2-3万/月" + "'" + "}," )
print("{value: " + str(d) + ", name: " + "'" + "3-4万/月" + "'" + "}," )
print("{value: " + str(e) + ", name: " + "'" + "4-5万/月" + "'" + "}," )
print("{value: " + str(f) + ", name: " + "'" + "5-6万/月" + "'" + "}," )
```
